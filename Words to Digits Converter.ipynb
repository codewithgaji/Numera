# Converts Words to Digits

```python
import time

def type_text(text, delay = 0.06):
    for char in (text):
        print(char, end = "", flush = True)
        time.sleep(delay)
    print()


def word_to_digits():
    type_text("Inspired by a friend :) ")
    base_num = {
        'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
        'six': 6, 'seven': 7, 'eight': 8, 'nine': 9
    }

    tens = {
        'ten': 10, 'twenty': 20, 'thirty': 30, 'forty': 40, 'fifty': 50,
        'sixty': 60, 'seventy': 70, 'eighty': 80, 'ninety': 90
    }

    hundreds = {'hundred': 100}
    thousands = {'thousand': 1000}
    millions = {'million' : 1000000}
    billions = {'billion': 1000000000}
    trillions = {'trillion': 1000000000000}

    while True:
        try:
            qs = input('What word number would you like to convert to digits? ').lower()
            qs_parts = qs.split()
            total = 0
            current_value = 0

            # Iterating throug the question parts 
            # e.g ('two hundred') splitted to ('two' 'hundred')
            for part in qs_parts:
                
                # This here checks if the iterated part is in base_num
                if part in base_num:
                    current_value += base_num[part]
                    
                # This here checks if the iterated part is in tens
                elif part in tens:
                    current_value += tens[part]
                    
                # This here checks if the iterated part is in hundreds
                # Instead of creating a dict of values like 200, 300,
                # instead I decided to use multiplication logic.
                # So we multiply the values like e.g 'two hundred' splitted to 
                # 'two' , 'hundred'. Once 'two' has been confirmed as
                # a digit '2' and added to the current_value, then we confirm 'hundred' 
                # as a digit '100' and multiply 2 by the value of 'hundred': 100
                # and we get our values
                elif part in hundreds:
                    current_value *= 100
                
                 # This here checks if the iterated part is in thousands
                # Same logic used in 'hundreds' is what is been 
                # used in 'thousands' elif statement
                # Just that after getting the total value for 'thousand' 
                # we return the 'current_value' back to '0'
                # The 'total += current_value' is so that it can add 
                # any remaining value from the input
                # So it can handle cases like 'two thousand five hundred' and
                # give values like '2000' + '500' after confirming the conditions
                elif part in thousands:
                    current_value *= 1000
                    total += current_value
                    current_value = 0
                    
                # The same logic used for thousands and 
                # hundreds is being implemented here
                elif part in millions:
                    current_value *= 1000000
                    total += current_value
                    org = f"{total:,}"
                    current_value = 0
                    
                # The same logic used for thousands and 
                # hundreds is being implemented here
                elif part in billions:
                    current_value *= 1000000000
                    total += current_value
                    org = f"{total:,}"
                    current_value = 0
                    
                # The same logic used for thousands and 
                # hundreds is being implemented here
                elif part in trillions: 
                    current_value *= 1000000000000
                    total += current_value
                    org = f"{total:,}"
                    current_value = 0
                    
                else:
                    print(f"\nUnknown word: {part}")
                    
                    
            # This right here give the correct output once
            # The calculations have been made
            total += current_value  # Add any remaining value
            if part in thousands or part in millions or part in billions or part in trillions:
                total += current_value
                org = f"{total:,}"   # This here allows easy number formatting 
                # the 'org' var is used to format something like:
                # 200000 to 200,000 by seperating it with commas ','
                current_value = 0  # Reset after handling thousands
                print(f"\nThe digit value is: {org}")
                    
            elif part not in thousands or part in millions or part in billions or part in trillions:
                    print(f"\nThe digit value is: {total}")
                    
        except ValueError:
            print('Please provide only a WORD NUMBER')


word_to_digits()
```
